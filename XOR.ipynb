```
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Activation
 
```

```
x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
y_train = np.array([[0],[1],[1],[0]])
```

```
 model = Sequential()

model.add(Dense(2,input_dim=2,activation='relu'))      # first hidden layer             
model.add(Dense(1, activation='sigmoid'))                # output layer
 
 
```

```
model.summary()
```

```
Weight_Input_Hidden = model.layers[0].get_weights()[0]
Weight_Output_layer = model.layers[1].get_weights()[0]

print('INPUT-HIDDEN LAYER WEIGHTS:')
print(Weight_Input_Hidden)
print('HIDDEN-OUTPUT LAYER WEIGHTS:')
print(Weight_Output_layer)
```

```
Bias_Input_Hidden = model.layers[0].get_weights()[1]
Bias_Output_layer = model.layers[1].get_weights()[1]
print(Bias_Input_Hidden)
print(Bias_Output_layer)
```

```
model.compile(optimizer='SGD', loss='MSE', metrics=['accuracy'])
```

```
model.fit(x_train, y_train,epochs=250)
```

```
x_test = np.array([[1,0]])
```

```
pred=model.predict(x_test).round() 
pred
```
